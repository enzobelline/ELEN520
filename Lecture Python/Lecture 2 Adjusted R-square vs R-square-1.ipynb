We use the example in this notebook to discuss the difference between adjusted R-squared and R-squared.

```python
import pandas as pd
# read the file Boston.csv into pandas dataframe
df = pd.read_csv('Cake.csv')
df.head() # view top portion of file
```

```python
import statsmodels.formula.api as smf
lm_fit = smf.ols('Cake~Dough', data=df).fit()
lm_fit.summary()
```

```python
import statsmodels.formula.api as smf
lm_fit = smf.ols('Cake~Dough+Sugar', data=df).fit()
lm_fit.summary()
```

```python
import statsmodels.formula.api as smf
# Note that when Temperature is added, R-squared increases, but Adj. R-squared decreases
lm_fit = smf.ols('Cake~Dough+Sugar+Temperature', data=df).fit()
lm_fit.summary()
```
