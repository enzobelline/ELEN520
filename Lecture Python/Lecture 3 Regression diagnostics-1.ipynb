```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np

df = pd.read_csv('Auto.csv').sort_values('horsepower')
lm_fit1 = smf.ols('mpg~horsepower', data = df).fit()
lm_fit2 = smf.ols('mpg~horsepower+np.power(horsepower, 2)', data = df).fit()

plt.scatter(df.horsepower, df.mpg)
y_pred1 = lm_fit1.predict(df.horsepower)
y_pred2 = lm_fit2.predict(df.horsepower)

plt.plot(df.horsepower, y_pred1, color="black")
plt.plot(df.horsepower, y_pred2, color="orange", linestyle="--", linewidth="3")

plt.title('Scatterplot of mpg vs horsepower in Auto.csv')
plt.ylabel('mpg')
plt.xlabel('horsepower')
```

```python
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np

df = pd.read_csv('Auto.csv').sort_values('horsepower')
lm_fit1 = smf.ols('mpg~horsepower', data = df).fit()
y_pred1 = lm_fit1.predict(df.horsepower)

# create scatter plot; color some observations 
plt.scatter(y_pred1,  df.mpg, color='black')
plt.scatter(y_pred1[100],  df.mpg[100], color='green')
plt.scatter(y_pred1[220],  df.mpg[220], color='orange')
plt.scatter(y_pred1[0],  df.mpg[0], color='red')

plt.title('Scatterplot of actual mpg vs predicted mpg')
plt.ylabel('actual mpg')
plt.xlabel('predicted mpg')
plt.ylim(0, 40)

# draw a line showing ideal case predicted mpg is equal to actual mpg
x1 = np.arange(40)
y1 = x1
b, m = np.polyfit(x1, y1, 1)
plt.plot(x1, b + m * x1, 'k-')
```

```python
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

df = pd.read_csv('Auto.csv').sort_values('horsepower')
lm_fit1 = smf.ols('mpg~horsepower', data = df).fit()
y_pred1 = lm_fit1.predict(df.horsepower)

plt.scatter(y_pred1,  df.mpg-y_pred1, color='black')
plt.scatter(y_pred1[100],  df.mpg[100]-y_pred1[100], color='green')
plt.scatter(y_pred1[220],  df.mpg[220]-y_pred1[220], color='orange')
plt.scatter(y_pred1[0],  df.mpg[0]-y_pred1[0], color='red')

plt.title('Scatterplot of residuals vs predicted mpg')
plt.ylabel('residuals')
plt.xlabel('predicted mpg')
```

```python
import statsmodels.api as sm
fig, ax = plt.subplots(figsize=(20,20))
# display an influence plot using Cook's distance
fig = sm.graphics.influence_plot(lm_fit1, ax=ax, criterion="cooks")
```

```python
import statsmodels.stats.outliers_influence 
olsi = statsmodels.stats.outliers_influence.OLSInfluence(lm_fit1)
olsi.summary_frame()
```
